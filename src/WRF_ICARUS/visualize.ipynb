```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ds_arr = []
dir = '.'
for fp in ['t2_500.nc']:
# dir = '06_05_13:00'
# for fp in ['t2_2000.nc', 't2_1000.nc', 't2_500.nc', 't2_250.nc', 't2_125.nc', 't2_62.nc']:
    dir = dir + '/' if dir[-1] != '/' else dir
    ds_arr.append(xr.open_dataset(dir + fp))
```

```python
for ds in ds_arr:
    plt.figure()
    t2 = ds.variables['T2'].values[0,:,:]
    x = np.arange(0,t2.shape[1]+1,1)
    y = np.arange(0,t2.shape[0]+1,1)
    plt.pcolormesh(x,y,t2)
    plt.colorbar()
    t = str(int(ds.DX)) + "m Resolution Output for " + dir.replace('/','')
    plt.title(t)
```
